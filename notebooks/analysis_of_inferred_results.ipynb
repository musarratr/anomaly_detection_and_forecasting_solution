```python
# %% Imports & plotting config

import numpy as np
import pandas as pd
import os
from pathlib import Path

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12, 5)
plt.rcParams["axes.grid"] = True

from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
DATA_DIR = os.path.abspath(os.path.join(os.getcwd(), "../" , "data"))
DATA_RAW_DIR = Path(os.path.join(DATA_DIR, "raw"))
DATA_PROCESSED_DIR = Path(os.path.join(DATA_DIR, "processed"))
# Path to the dataset (adjust this if needed)
inferred_path = Path(os.path.join(DATA_PROCESSED_DIR, "oxygen_processed_full.csv"))
```

```python
df = pd.read_csv(inferred_path, index_col=0)
```

```python
df = df.reset_index()
```

```python
df = df[:10000]
```

```python
def plot_two_dotted_lines(df, x_col, y1_col, y2_col):
    """
    Plot two dotted line graphs for two continuous columns in the same DataFrame.
    
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame containing the data.
    x_col : str
        Column name to use on the x-axis.
    y1_col : str
        First continuous column to plot.
    y2_col : str
        Second continuous column to plot.
    """
    plt.figure(figsize=(10, 5))

    # Dotted lines with linestyle=":"
    plt.plot(df[x_col], df[y1_col], linestyle=":", label=y1_col)
    plt.plot(df[x_col], df[y2_col], linestyle=":", label=y2_col)

    plt.xlabel(x_col)
    plt.ylabel("Value")
    plt.title("Two Dotted Line Graphs")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_two_dotted_lines(df, x_col="time", y1_col="Oxygen[%sat]", y2_col="y_pred")
```
