```python
import filecmp
import os
from pathlib import Path

def csv_files_equal_exact(path1: str, path2: str) -> bool:
    """
    Return True if the two CSV files are exactly the same (byte-for-byte).
    """
    return filecmp.cmp(path1, path2, shallow=False)
```

```python
DATA_DIR = os.path.abspath(os.path.join(os.getcwd(), "../" , "data"))
DATA_RAW_DIR = Path(os.path.join(DATA_DIR, "raw"))
DATA_PROCESSED_DIR = Path(os.path.join(DATA_DIR, "processed"))
# Path to the dataset (adjust this if needed)
f1 = Path(os.path.join(DATA_PROCESSED_DIR, "oxygen_processed_sample_1.csv"))
f2 = Path(os.path.join(DATA_PROCESSED_DIR, "oxygen_processed_sample_2.csv"))
```

```python
csv_files_equal_exact(f1, f2)
```
