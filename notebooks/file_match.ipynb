```python
import filecmp

def csv_files_equal_exact(path1: str, path2: str) -> bool:
    """
    Return True if the two CSV files are exactly the same (byte-for-byte).
    """
    return filecmp.cmp(path1, path2, shallow=False)
```

```python
f1 = "scored_v1.csv"
f2 = "scored_v2.csv"

csv_files_equal_exact(f1, f2)
```
